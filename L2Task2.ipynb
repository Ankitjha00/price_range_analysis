```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Dataset .csv")
df.head()
```

```python
price_counts = df['Price range'].value_counts()
price_counts
```

```python
avg_rating_price = df.groupby('Price range')['Aggregate rating'].mean()
avg_rating_price
```

```python
plt.figure(figsize=(8,5))
bars = plt.bar(avg_rating_price.index, avg_rating_price.values)

# Highlight highest average rating
max_index = avg_rating_price.idxmax()
bars[list(avg_rating_price.index).index(max_index)].set_color('green')

plt.title("Average Rating by Price Range")
plt.xlabel("Price Range")
plt.ylabel("Average Rating")
plt.show()
```
